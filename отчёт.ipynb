```python
import pandas as pd

def calculate_text_stats(series):
    """
    Возвращает статистику по текстовым данным в серии:
    - минимальная длина строки (символов)
    - максимальная длина строки (символов)
    - средняя длина строки (символов)
    - количество слов
    """
    # Проверяем, что серия содержит текстовые данные
    if not pd.api.types.is_string_dtype(series.dtype):
        return {
            'min_length': None,
            'max_length': None,
            'mean_length': None,
            'min_words': None,
            'max_words': None,
            'mean_words': None
        }

    # Заполняем NaN пустыми строками, чтобы избежать ошибок с len()
    series = series.fillna('')

    # Вычисляем длину строк (включая пробелы)
    length = series.apply(len)

    # Вычисляем количество слов (разделение по пробелам)
    word_count = series.str.split().str.len()

    return {
        'min_length': length.min(),
        'max_length': length.max(),
        'mean_length': length.mean(),
        'min_words': word_count.min(),
        'max_words': word_count.max(),
        'mean_words': word_count.mean()
    }

def analyze_text_data(df):
    """
    Анализ текстовых колонок в DataFrame и возвращает результат в виде DataFrame.
    """
    # Выбираем только текстовые столбцы (тип object)
    text_cols = df.select_dtypes(include='object').columns

    if not text_cols.empty:
        results = []
        for col in text_cols:
            stats = calculate_text_stats(df[col])
            results.append(stats)

        # Создаем DataFrame с результатами
        result_df = pd.DataFrame(results, index=text_cols)
        return result_df
    else:
        return "Нет текстовых колонок в DataFrame"
```
